# Stationarity

## Weakly Stationary Examples
- (Gaussian) white noise
- Moving average process, MA($q$)

## Intuition and Definition
We say a stochastic process is **strictly stationary** if joint distribution of $X(t_1), X(t_2), \dots, X(t_k)$ is the same as the joint distribution of $X(t_1+\tau), X(t_2+\tau), \dots, X(t_k+\tau)$.

The random variables are identically distributed but not necessarily independent.

In particular, the joint distribution *only depends upon our lag spacings*.

- Constant mean and variance
- Autocovariance function: $\gamma(t_1,t_2) = \gamma(t_2-t_1) = \gamma(\tau)$ only depends on separation
- Autocorrelation function: $\rho(\tau) \equiv \dfrac{\gamma(\tau)}{\gamma(0)}$


We say a stochastic process is **weakly stationary** if the mean function is constant, or, Autocovariance function (acf) $\gamma(t_1,t_2) = \gamma(t_2-t_1) = \gamma(\tau)$ only depends on separation.

- Implication: constant variance


## Develop the ACF of a moving average process
- $Z_t\sim iid(0,\sigma^2)$
- MA(q) process: $X_t = \beta_0Z_t + \beta_1Z_{t-1} +\dots+ \beta_qZ_{t-q}$

Because of the zero mean, $Cov[X,Y] = E[XY]$. Then

$$Cov[X_t, X_{t+k}] = E[(\beta_0Z_t+\dots+\beta_qZ_{t-q})(\beta_0Z_{t+k}+\dots+\beta_qZ_{t+k-q})]$$

Notice $E[Z_iZ_j] = 0 \text{ if } i\ne j \text{ else } \sigma^2$, when $k\le q$

$$Cov[X_t, X_{t+k}] = \sigma^2 \sum_{i=0}^{q-k} \beta_i\beta_{i+k}$$ 

not $t$ dependent.

# Backward Shift Operator $B$

$BX_t = X_{t-1}$

$B^kX_t = X_{t-k}$

## Examples

- Random walk $X_t = X_{t-1}+Z_t$

  $(1-B)X_t=Z_t$


- MA(2) process $X_t = Z_t + 0.2Z_{t-1} + 0.04𝑍_{t-2}$
  
  $X_t = (1+0.2B+0.04B^2)Z_t$
 
 
- AR(2) process $X_t = 0.2X_{t-1} + 0.3X_{t-2}+Z_t$
  
  $(1-0.2B-0.3B^2)X_t = Z_t$
  
  
- MA(q) process with a drift $X_t = \mu + \beta_0Z_t + \beta_1Z_{t-1} +\dots+ \beta_qZ_{t-q}$
  
  $X_t - \mu = \beta(B)Z_t$
  
  where $\beta(B) = \beta_0 + \beta_1B +\dots+\beta_q B^q$
  
  
- AR(p) process $X_t = \phi_1X_{t-1} + \phi_2X_{t-2} +\dots + \phi_pX_{t-p} +Z_t$
  
  $\phi(B)X_t = Z_t$
  
  where $\phi(B) = 1 - \phi_1B - \phi_2B^2 -\dots - \phi_pB^p$
  
  
# Invertibility
## Inverting using backward substitution
MA(1) process: $X_t = Z_t+\beta Z_{t-1}$

$Z_t = X_t - \beta Z_{t-1} = X_t - \beta (X_{t-1} -\beta Z_{t-2}) = \dots$

In this manner,

$Z_t = X_t - \beta X_{t-1} + \beta^2 X_{t-2} -\beta^3 X_{t-3} +\dots$

i.e.,

$X_t = Z_t + \beta X_{t-1} - \beta^2 X_{t-2} +\beta^3 X_{t-3} -\dots$

We inverted MA(1) to AR($\infty$).

## Inverting using backward shift operator
MA(1) process: $X_t = Z_t+\beta Z_{t-1} = \beta(B)Z_t$, where $\beta(B)=1+\beta B$.

Then we find $Z_t$ by inverting the polynomial operator $\beta(B)$: $\beta^{-1}(B)X_t = Z_t$. And

$$\beta^{-1}(B) = \frac{1}{1+\beta B} = 1 - \beta B + \beta^2 B^2 - \beta^3 B^3 + \dots$$

Hence, $$Z_t = \beta^{-1}(B)X_t = X_t - \beta^2 X_{t-2} +\beta^3 X_{t-3} -\dots = \sum_{n=0}^\infty \beta^n X_{t-n}$$


## Definition
$\{X_t\}$ is called **invertible** if $Z_t = \sum_{k=0}^\infty \pi_k X_{t-k}$ where $\sum_{k=0}^\infty |\pi_k|$ is convergent.


# Duality
## Invertibility condition for MA(q) process
MA(q) process:

$$X_t = \beta_0Z_t + \beta_1Z_{t-1} +\dots+ \beta_qZ_{t-q} = (\beta_0 + \beta_1 B + \dots + \beta_qB^q)Z_t = \beta(B)Z_t$$

We obtain innovations $Z_t$ in terms of present and past values of $X_t$,

$$Z_t = \beta^{-1}(B)X_t = (\alpha_0 + \alpha_1B+\alpha_2B^2+ \dots )X_t$$

We say MA(q) process is **invertible** if the roots of the polynomial $\beta(B)$ all lie outside the unit circle, where we regard $B$ as a complex variable (not an operator).

(Proof is done using mean-square convergence)

### Examples
1. MA(1) process $X_t = Z_t+\beta Z_{t-1}$
  
  $\beta(B)=1+\beta B = 0 \implies B = -1/\beta$ (only root)
  
  $|-1/\beta|>1 \implies |\beta|<1$
  
  Then, $Z_t = \sum_{k=0}^\infty (-\beta)^kB^kX_t = \sum_{k=0}^\infty (-\beta)^kX_{t-k}$
  
  
2. MA(2) process $X_t = Z_t + \frac{5}{6}Z_{t-1} + \frac{1}{6}Z_{t-2}$
  
  $\beta(B) = 1 + \frac{5}{6}B + \frac{1}{6}B^2 = 0 \implies B_1 = 2, B_2=3$ (real numbers)
  
  $$\beta(B)^{-1} = \frac{1}{1 + \frac{5}{6}B + \frac{1}{6}B^2} = \frac{3}{1+\frac{1}{2}B} - \frac{2}{1+\frac{1}{3}B}$$
  
  Expand both fractions using geometric series,
  
  $$\beta(B)^{-1} = \sum_{k=0}^\infty [3(-\frac{1}{2})^k - 2(-\frac{1}{3})^k]B^k$$
  
  MA(2) $\implies$ AR($\infty$)

## Stationarity condition for AR(p) process

MA(q) process is always stationary but AR(p) is not. AR(p) process is **(weakly) stationary** if the root of the polynomial 

$$\phi(B) = 1 - \phi_1B - \phi_2B^2 -\dots - \phi_pB^p$$

all lie outside the unit circle, where we regard $B$ as a complex variable (not an operator).

### AR(1) process

AR(p) $\implies$ MA($\infty$)

## Duality

Under invertibility condition of MA(q), MA(q) $\implies$ AR($\infty$).

Under stationarity condition of AR(p), AR(p) $\implies$ MA($\infty$).

# Autoregressive Process AR(p)

## Simulate an AR(1)

```r
set.seed(2016); N=1000; phi=.4;
Z = rnorm(1000); X=NULL;
X[1] = Z[1];

for (t in 2:N){
    X[t] = Z[t] + phi*X[t-1];
}

X.ts = ts(X)

par(mfrow=c(2,1))
plot(X.ts, main='AR(1) Time Series on White Noise, phi=.4')
X.acf = acf(X.ts, main='AR(1) Time Series on White Noise, phi=.4')
```

## Simulate an AR(2)
$X_t = Z_t + .7X_{t-1} + .2X_{t-2}$

```r
X.ts <- arima.sim(list(ar = c(.7, .2)), n=1000)
par(mfrow=c(2,1))
plot(X.ts,main="AR(1) Time Series, phi1=.7, phi2=.2")
X.acf = acf(X.ts, main="Autocorrelation of AR(1) Time Series")
# correlation decays at a slower rate now
```

```r
# Another AR(2)
phi1 = .5; phi2 = -.4;
X.ts <- arima.sim(list(ar = c(phi1, phi2)), n=1000)
par(mfrow=c(2,1))
plot(X.ts,main=paste("AR(1) Time Series, phi1=", phi1, "phi2=", phi2))
X.acf = acf(X.ts, main="Autocorrelation of AR(1) Time Series")
```

## Backshift Operator and the ACF
From above we know, AR(p) process $X_t = \phi_1X_{t-1} + \phi_2X_{t-2} +\dots + \phi_pX_{t-p} +Z_t$
  
  $\phi(B)X_t = Z_t$
  
  where $\phi(B) = 1 - \phi_1B - \phi_2B^2 -\dots - \phi_pB^p$.
  
We can write (geometric series expansion)

$$X_t = \frac{1}{1 - (\phi_1B + \phi_2B^2 +\dots + \phi_pB^p)}Z_t = (1+\theta_1B+\theta_2B^2 +\dots)Z_t$$

### Results

- Expected value:
  $E[X_t] = E[Z_t] + \theta_1E[Z_{t-1}] + \theta_2E[Z_{t-2}] + \dots = 0$
  
  
- Variance:
  $V[X_t] = V[Z_t] + \theta_1^2E[Z_{t-1}] + \theta_2^2E[Z_{t-2}] + \dots = \sigma_z^2 \sum_{i=0}^\infty \theta_i^2$
  
  Necessary condition for stationarity: the sum must converge.
  
### Autocovariance

For a MA(q) process, $\gamma(k) = \sigma_z^2\cdot \sum_{i=0}^{q-k} \theta_i \theta_{i+k}$.

For a AR(p) process, $\gamma(k) = \sigma_z^2\cdot \sum_{i=0}^\infty \theta_i \theta_{i+k}$.

### Autocorrelation

For a AR(p) process,

$$\rho(k) = \frac{\gamma(k)}{\gamma(0)} = \frac{\sum_{i=0}^\infty \theta_i \theta_{i+k}}{\sum_{i=0}^\infty \theta_i^2}$$

# Yule-Walker Equations

## Difference equation

Given $a_n = 5a_{n-1} - 6a_{-2}$, find $a_n$.

_Solution:_ Set $a_n=\lambda^n$, 

$\lambda^n = 5\lambda^{n-1} - 6 \lambda^{n-2}$ (auxilary equation / characteristic equation), i.e., $\lambda^2 -5\lambda+6=0 \implies \lambda=2,3$

$a_n = c_12^n+c_23^n$ ...

### $k$-th order difference equation

## Yule-Walker Equations

### Procedure
- We assume stationarity in advance (a priori assumption)
- Take product of the AR model with $X_{n-k}$
- Take expectation of both sides
- Use the definition of covariance, and divide by $\gamma(0)=\sigma_X^2$
- Get difference equation for $\rho(k)$, ACF of the process
- This set of equations is called Yule-Walker equations
- Solve the difference equation

### Example
AR(2) process $X_t = \frac{1}{3}X_{t-1} + \frac{1}{2}X_{t-2} + Z_t$

Polynomial $\phi(B) = 1-\frac{1}{3}B-\frac{1}{2}B^2$ has real roots $\frac{-2\pm\sqrt{76}}{6}$ both of which has magnitude greater than 1. Thus it is stationary.
